# 新聞記事の単語解析

predict値が0.7より多くさがった5文字を抽出

```python
import pickle

def save_words(filepath, word_list):
    with open(filepath , 'rb') as f:
        datas = pickle.load(f)
        for data in datas:
            if data == '':
                pass
            else:
                data = data.split('\n')
                for word in data:
                    if word == '':
                        pass
                    else:
                        word.replace('\n', '')
                        word_list.append(word)
```

```python
yomiuri_words = []
save_words('yomiuri_five_words.pkl', yomiuri_words)
```

```python
asahi_words = []
save_words('asahi_five_words.pkl', asahi_words)
```

```python
mainichi_words = []
save_words('mainichi_five_words.pkl', mainichi_words)
```

```python
sankei_words = []
save_words('sankei_five_words.pkl', sankei_words)
```
