```python
files = [ 'yomiuri_words.pkl', 'asahi_words.pkl', 'mainichi_words.pkl', 'sankei_words.pkl' ]
words = [ 'yomiuri_words', 'asahi_words', 'mainichi_words', 'sankei_words' ]

def save_words(filepath, word_list):
    with open(filepath , 'rb') as f:
        tmp = pickle.load(f)
    ex_words = []
    for w in tmp:
        if w == '':
            pass
        else:
            ex_words.append(w)
    for w in ex_words:
        ex_w = w.split('\n')
        for e in ex_w:
            if e == '':
                pass
            else:
                e = e.replace('\x00', '')
                word_list.append(e)
```

```python
yomiuri_words = []
save_words('yomiuri_words.pkl', yomiuri_words)
```

```python
mainichi_words = []
save_words('mainichi_words.pkl', mainichi_words)
```
