# 新聞記事の単語解析

基記事とマスクした記事との出力の差を影響度としてラスタスキャンし合計値を求める。
影響度の合計を0~1に正規化し、その値が0.5を超える部位を抽出。

```python
import pickle

def save_words(filepath, word_list):
    with open(filepath , 'rb') as f:
        datas = pickle.load(f)
        for data in datas:
            if data == '':
                pass
            else:
                data = data.split('\n')
                for word in data:
                    if word == '':
                        pass
                    else:
                        word.replace('\n', '')
                        word_list.append(word)
```

```python
yomiuri_words = []
save_words('yomiuri_words.pkl', yomiuri_words)
```

```python
asahi_words = []
save_words('asahi_words.pkl', asahi_words)
```

```python
mainichi_words = []
save_words('mainichi_words.pkl', mainichi_words)
```

```python
sankei_words = []
save_words('sankei_words.pkl', sankei_words)
```
